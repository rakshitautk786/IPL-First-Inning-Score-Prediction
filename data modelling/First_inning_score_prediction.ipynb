# First Inning Score Prediction

```python
#importing essential libraries
import pandas as pd
import pickle


#loading the dataset
df=pd.read_csv('ipl.csv')
```

```python
df.head()
```

```python
#--data cleaning--
#removing unwanted columns
columns_to_remove=['mid','venue','batsman','bowler','striker','non-striker']
df.drop(labels=columns_to_remove,axis=1,inplace=True)
```

```python
df['bat_team'].unique()
```

```python
df.head()
```

```python
df['bat_team'].unique()#how many unique team
```

```python
#keep only consistent teams(only teams which are playing)

consistent_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Deccan Chargers', 'Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils',
       'Kochi Tuskers Kerala', 'Pune Warriors', 'Sunrisers Hyderabad',
       'Rising Pune Supergiants', 'Gujarat Lions',
       'Rising Pune Supergiant']
```

```python
# filtering consistent team from datasert
df=df[(df['bat_team'].isin(consistent_teams))&(df['bowl_team'].isin(consistent_teams))]
```

```python
#Removing the first 5 overs data in every match
#u need to have atleast min 5 hour data to predict the score

df=df[df['overs']>=5.0]
```

```python
df.head()
```

```python
print(df['bat_team'].unique())
print(df['bowl_team'].unique())
```

```python
#coverting date col. of string to date time object
#string to datetime obj 
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
df.info()
```

```python
#--Data Processing ---\
#converting categorical features using one hot encoding
encoded_df=pd.get_dummies(data=df,columns=['bat_team','bowl_team'])
pd.pandas.set_option('display.max_columns', None)
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

```python
#Rearranging the columns
# to train and test data putting total at last
encoded_df=encoded_df[['date', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5',
        'bat_team_Chennai Super Kings', 'bat_team_Deccan Chargers',
       'bat_team_Delhi Daredevils', 'bat_team_Gujarat Lions',
       'bat_team_Kings XI Punjab', 'bat_team_Kochi Tuskers Kerala',
       'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians',
       'bat_team_Pune Warriors', 'bat_team_Rajasthan Royals',
       'bat_team_Rising Pune Supergiant', 'bat_team_Rising Pune Supergiants',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Deccan Chargers',
       'bowl_team_Delhi Daredevils', 'bowl_team_Gujarat Lions',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kochi Tuskers Kerala',
       'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians',
       'bowl_team_Pune Warriors', 'bowl_team_Rajasthan Royals',
       'bowl_team_Rising Pune Supergiant', 'bowl_team_Rising Pune Supergiants',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','total']]
```

```python
encoded_df.info()
```

```python
#splitting  the data into train and test data
#time based splitting
x_train= encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year<=2016]
x_test= encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year>=2017]
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train=encoded_df[encoded_df['date'].dt.year<=2016]['total'].values
y_test=encoded_df[encoded_df['date'].dt.year>=2017]['total'].values
```

```python
# date time only use to train and test split so drop it
#Removing the date column
x_train.drop(labels='date',axis=True,inplace=True)
x_test.drop(labels='date',axis=True,inplace=True)
```

```python
x_train.info()
```

```python
#--Model Building
#linear Regression model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#creating a pickle file for the classifier

filename='first_innings_lr_model.pkl'
pickle.dump(regressor,open(filename,'wb'))
```

# ridge Regression

```python
##ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(x_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
prediction = ridge_regressor.predict(x_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
% inline matplotlib

sns.distplot(y_test-prediction)
plt.show()
```

```python
from sklearn import metrics
import numpy as np
print('MAE:',metrics.mean_absolute_error(y_test,prediction))
print('MSE:',metrics.mean_squared_error(y_test,prediction))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

# Lasso Regression

```python
# Lasso is slow

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(x_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
prediction = lasso_regressor.predict(x_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(y_test-prediction)
plt.show()
```

```python
from sklearn import metrics
import numpy as np
print('MAE:',metrics.mean_absolute_error(y_test,prediction))
print('MSE:',metrics.mean_squared_error(y_test,prediction))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```
